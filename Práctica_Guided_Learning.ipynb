```python
# Creamos un nuevo sub-conjunto de datos solo con el año 2025
datos_2025 = df[df['Year'] == 2025]

# Comprobamos cuántos registros hay en este grupo
print(f"Total de publicaciones en 2025: {len(datos_2025)}")

# Vemos los primeros 10 títulos para buscar patrones a simple vista
datos_2025['Title'].head(10)
```

```python
from collections import Counter
import pandas as pd

# 1. Filtramos los datos para quedarnos solo con el año 2025
df_2025 = df[df['Year'] == 2025]

# 2. Analizamos los Títulos: Extraemos todas las palabras
# Combinamos los títulos, pasamos a minúsculas y separamos por espacios
palabras = " ".join(df_2025['Title'].astype(str)).lower().split()

# 3. Filtramos "palabras vacías" (opcional pero recomendado)
# Esto evita que aparezcan palabras como 'of', 'the', 'and', 'in'
stop_words = {'of', 'the', 'and', 'in', 'on', 'with', 'for', 'a', 'to', 'from'}
palabras_limpias = [p for p in palabras if p not in stop_words and len(p) > 3]

# 4. Contamos las 15 palabras más comunes
top_temas = Counter(palabras_limpias).most_common(15)

print("--- Temas más frecuentes en 2025 ---")
for palabra, count in top_temas:
    print(f"{palabra.capitalize()}: {count} menciones")

print("\n--- Revistas/Fuentes con más publicaciones en 2025 ---")
print(df_2025['Source title'].value_counts().head(5))
```

```python
# 1. Tomamos los datos de 2025
df_2025 = df[df['Year'] == 2025].copy()

# 2. Limpiamos y separamos los autores
# Convertimos a string, quitamos espacios y separamos por ';'
autores_separados = df_2025['Authors'].astype(str).str.split(';')

# 3. "Explotamos" la lista para que cada autor sea una fila independiente
autores_series = autores_separados.explode().str.strip()

# 4. Contamos los 10 autores más frecuentes
top_autores = autores_series.value_counts().head(10)

print("--- Autores líderes en 2025 ---")
print(top_autores)
```
